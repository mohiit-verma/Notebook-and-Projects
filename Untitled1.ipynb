```python
# function to combine values of the important column into a single string

def imp_feature(data):
    imp_f = []
    for i in range(data.shape[0]):
        s = df['director_name'][i] + ' ' + df['actor_2_name'][i] + ' ' +  df['actor_1_name'][i] + ' ' \
                     + df['actor_3_name'][i] + ' ' + df['genres'][i] + ' ' + df['language'][i] + ' ' + df['country'][i]
        imp_f.append(s)

    return imp_f
```

```python
df['important_features'] = imp_feature(df)
```

```python
df = df.dropna()
df.info()
```

```python

#df['important_features'].fillna('xyz')
#df['important_features'] = df['important_features'].str.lower()
df = df.convert_dtypes()
```

```python
cm = CountVectorizer().fit_transform(df['important_features'])
```

```python
cs = cosine_similarity(cm)
print(cs)
cs.shape
```

```python
title = 'The Dark Knight Rises'
movie_id = df[df['movie_title'] == title].index
#movie_id = df[df.movie_title == title]['movie_id'].values[0]
```

```python
movie_id
```

```python
scores = list(enumerate(cs[1]))
df.iloc[1]
```

```python
sorted_scores = sorted(scores, key = lambda x:x[1], reverse = True)
sorted_scores = sorted_scores[1:]
```

```python
#print(sorted_scores, type(sorted_scores))
```

```python
j = 0
df_result = pd.DataFrame()
title = []
genre = []
sc = []
for i in sorted_scores:
    
    movie_title = df[df.movie_id == i[0]]['movie_title'].values[0]
    genres = df[df.movie_id == i[0]]['genres'].values[0]
    title.append(movie_title)
    genre.append(genres)
    sc.append(i[1])
    #print(df_result)
    j += 1
    if j > 20:
        break
df_result['Recommended_Movie'] = title
df_result['Genre'] = genre
df_result['Score'] = sc
df_result
```
